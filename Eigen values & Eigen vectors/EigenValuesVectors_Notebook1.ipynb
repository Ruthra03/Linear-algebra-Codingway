##Problem Statement

Let A=[[1,6],[5,2]],

u=[6,-5],
v=[3,-2]

Are u and v eigen vectors of A? Prove that visually.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the matrix A
A = np.array([[1, 6],[5, 2]])
# Define the vectors u and v
u = np.array([6, -5])
v = np.array([3, -2])
```

```python
#Calculate end points after tranformation
Au = np.dot(A, u)
Av = np.dot(A, v)
```

```python
# Plot the vectors
plt.figure(figsize=(8, 6))

# Plot vector u in red with label 'u'
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b', label='u')

# Plot vector v in blue with label 'v'
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='y', label='v')

# Plot transformed vector Au in green with label 'Au'
plt.quiver(0, 0, Au[0], Au[1], angles='xy', scale_units='xy', scale=1, color='c', label='Au')

# Plot transformed vector Av in yellow with label 'Av'
plt.quiver(0, 0, Av[0], Av[1], angles='xy', scale_units='xy', scale=1, color='m', label='Av')

# Set x and y limits for the plot
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Add horizontal and vertical lines at x=0 and y=0
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add grid lines
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Add legend to the plot
plt.legend()

# Set title and axis labels
plt.title('Visualization of vectors u, v and their transformations by matrix A')
plt.xlabel('X')
plt.ylabel('Y')

# Display the plot
plt.show()
```
